```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
autos2 = pd.read_csv("autos_corregidos.csv")
autos2.head()
```

REGRESION LINEAL SIMPLE ( UNA VARIABLE)

```python
from sklearn.linear_model import LinearRegression
from sklearn import model_selection
```

```python
lm = LinearRegression()

#variables independientes o variables predictoras
x = autos2[["city-mpg"]]

# Variable dependiente o variable objetivo a predecir
y = autos2[["price"]]
x.head(10)
```

```python
#Entrenar el algoritmo de regresion lineal
lm.fit(x, y)
```

```python
#Realizamos la fase de prediccion
Ypredict = lm.predict(x)
```

```python
#Valor de la pendiente
lm.intercept_
```

```python
# valor del coeficiente o peso de la variable
lm.coef_
```

```python
lm.score(x,y)
```

##REGRESION LINEAL MULTIPLE


```python
autos2.info()
```

```python
# Variables predictorias para la fase de entrenamiento
```

```python
from sklearn.model_selection import train_test_split

# dividir el dataset en 2 partes, una par entrenamiento de datos y otra para probar
# Para machine learning es necesario dividir los datos en 2 partes, 1 parte
# aproximadamente del 70% de los datos, y loa otra del 30%
# la primera parte para entrenar el algoritmo de regresion lineal y la otra para probar
# la precision del algoritmo
autos2_train, autos2_test = train_test_split(autos2, random_state = 0)
```

#Regresion lineal multiple

```python
# ver la correlacion entre variable
autos2[["horsepower", "engine-size", "city-mpg", "wheel-base", "curb-weight", "price"]].corr()
```

```python
# Variables predictorias para la fase de entrenamiento
X_autos2_train = autos2_train[["horsepower", "engine-size", "city-mpg", "wheel-base", "bore"]]

# Variable a predicir fase de entrenamiento
y_autos2_train = autos2_train[["price"]]
```

```python
#Variables predictoras para la fase de prueba
X_autos2_test = autos2_test[["horsepower", "engine-size", "city-mpg", "wheel-base", "bore"]]

# Variable a predecir fase de prueba
y_autos2_test = autos2_test[["price"]]
```

```python
# Fase de entrenamineto del modelo lineal lm METODO FIT
lm.fit(X_autos2_train, y_autos2_train)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.score(X_autos2_train, y_autos2_train)
```

##FASE DE PRUEBA

```python
# Fase de prueba del modelo lineal lm usando el metodo PREDICT
yhat=lm.predict( X_autos2_test)
```

```python
lm.score( X_autos2_test, y_autos2_test)
```

```python
import seaborn as sns
plt.figure(figsize = (6,4))

ax1 = sns.distplot( y_autos2_test["price"], hist=False, color = "r")
ax1 = sns.distplot (yhat, hist=False, color= "b", ax = ax1)

plt.title("Precio real vs precio prediccion")
ax1.legend(labels=["Valor Real", "Valor Prediccion"])

plt.show()
plt.close()
```

```python
plt.figure(figsize = (6,4))

ax1 = sns.kdeplot( y_autos2_test["price"], multiple="stack", alpha=0.5, color = "r")
ax1 = sns.kdeplot ( pd.Series(yhat.flatten()),multiple="stack", alpha=0.5, color = "r")

plt.title("Precio real vs precio prediccion")
ax1.legend(labels=["Valor Real", "Valor Prediccion"])

plt.show()
plt.close()
```
