```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
import statistics as stats
import numpy as np
```

```python
#Cargar el archivo con una variable llamada autos
autos= pd.read_csv("autos.csv", encoding= "ISO-8859-1")
```

```python
#Mostrar los datos estadisticos descriptivos de precios
print("Estadísticas descriptivas:")
autos.price.describe()
```

```python
#Mostrar la moda de la base de datos de precios
moda= stats.mode(autos.price)
print(moda)
```

```python
#Mostrar la media de la base de datos de precios
print("Media:")
autos.price.mean()
```

```python
#Mostrar la mediana de la base de datos de precios
print("Mediana:")
autos.price.median()
```

```python
#Mostrar la desviacion estandar de la base de datos de precios
print("Desviacion Estandar:")
autos.price.std()
```

```python
#Histograma
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np

# Calcular estadísticas
media = autos.price.mean()
mediana = autos.price.median()
desviacion = autos.price.std()

# Graficar el histograma
plt.hist(autos.price, bins=11, alpha=0.2, color='b', edgecolor="black")

# Agregar líneas de estadísticos
plt.axvline(media, color='b', linestyle='dashed', linewidth=2, label=f'Media: {media:.2f}')
plt.axvline(mediana, color='r', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana:.2f}')
plt.axvline(moda, color='y', linestyle='dashed', linewidth=2, label=f'Moda: {moda:.2f}')
plt.axvline(desviacion, color='purple', linestyle='dashed', linewidth=2, label=f'Desviación: {desviacion:.2f}')

# Configuración del gráfico
plt.legend()
plt.title("Histograma de Precios de Autos")
plt.xlabel('Precio')
plt.ylabel('Frecuencias')
```

##Conclusiones:
La mayoría de los autos tienen precios más bajos, concentrándose en el rango de 5,000 a $15,000.

Existen algunos autos con precios mucho más altos, lo que genera una cola larga hacia la derecha.

Moda $5,572: Es el precio más frecuente y el más bajo de los tres valores.

Mediana ($10,295): La mitad de los autos cuestan menos de este valor.

Media ($13,207): Es mayor que la mediana debido a autos más caros que elevan el promedio.

La desviación estándar ($7,947) muestra que hay una variabilidad considerable en los precios.

```python
# Contar frecuencia de cada categoría
drive_counts = autos['body-style'].value_counts()
# Crear gráfico de barras
plt.figure(figsize=(10,5))
plt.bar(drive_counts.index, drive_counts.values, color='y',linestyle='dashed')

# Etiquetas y título
plt.title("Distribución de Tipos de Vehiculos (body style)")
plt.xlabel("Tipo de Vehiculos")
plt.ylabel("Precio")
```

```python
# Contar frecuencia de cada categoría
drive_counts = autos['drive-wheels'].value_counts()
# Crear gráfico de barras
plt.figure(figsize=(10,10))
plt.bar(drive_counts.index, drive_counts.values, color='y',linestyle='dashed')

# Etiquetas y título
plt.title("Distribución de Tipos de Tracción (drive wheels)")
plt.xlabel("Tipo de Tracción")
plt.ylabel("Precio")
```

```python
# Contar frecuencia de cada categoría
drive_counts = autos['engine-location'].value_counts()
# Crear gráfico de barras
plt.figure(figsize=(7,7))
plt.bar(drive_counts.index, drive_counts.values, color='y',linestyle='dashed')

# Etiquetas y título
plt.title("Distribución de Ubicacion del Motor (engine location)")
plt.xlabel("Ubicacion del Motor")
plt.ylabel("Precio")
```

```python
#Diagrama Caja y Bigotes entre Precio y Tipo de Vehiculo
sns.boxplot(x="price", y="body-style", data=autos)

plt.title("Diagrama de Caja: Precio vs Tipo de Vehículo")
plt.xlabel("Precio")
plt.ylabel("Tipo de Vehículo")
```

##Conclusiones:
Los precios de los vehículos varían según el tipo, con los convertibles y hardtops mostrando mayor dispersión y costos más elevados en general. Mientras tanto, los hatchbacks y wagons son más económicos y tienen precios más consistentes.

```python
#Diagrama Caja y Bigotes entre Precio y Tipo de Traccion
sns.boxplot(x="price", y="drive-wheels", data=autos)

plt.title("Diagrama de Caja: Precio vs Tipo de Traccion")
plt.xlabel("Precio")
plt.ylabel("Tipo de Traccion")
```

##Conclusiones:
Los vehículos con tracción trasera (rwd) tienden a ser los más costosos y presentan la mayor variabilidad de precios. Los de tracción delantera (fwd) son generalmente más asequibles y tienen una distribución más homogénea. Los de tracción en las cuatro ruedas (4wd) presentan un rango de precios más estrecho, lo que sugiere menor diversidad en los modelos disponibles en esta categoría.

```python
#Diagrama Caja y Bigotes entre Precio y Ubicacion del Motor
sns.boxplot(x="price", y="engine-location", data=autos)

plt.title("Diagrama de Caja: Precio vs Ubicacion del Motor")
plt.xlabel("Precio")
plt.ylabel("Ubicacion del Motor")
```

##Conclusiones:
Los vehículos con motor delantero son más comunes y tienen una amplia variabilidad en precios, desde modelos económicos hasta algunos más costosos. En cambio, los vehículos con motor trasero tienden a ser más exclusivos y de alto costo, con menos variabilidad en los precios. Esto concuerda con la tendencia de que los autos deportivos y de lujo suelen tener el motor en la parte trasera.
