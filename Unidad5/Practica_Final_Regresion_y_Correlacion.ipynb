```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
import os
```

```python

def analizar_grupo(nombre_archivo, nombre_grupo):
    print(f"\n--- Anlisis de {nombre_grupo} ---")

    df = pd.read_csv(nombre_archivo)

    print("\nPrimeras filas del DataFrame:")
    print(df.head())

    print("\nEstadsticas descriptivas:")
    print(df.describe())
    plt.figure(figsize=(6,4))
    sns.histplot(df['Calificacion'], kde=True, color='skyblue')
    plt.title(f'Histograma de Calificaciones - {nombre_grupo}')
    plt.xlabel('Calificacin Final')
    plt.ylabel('Frecuencia')
    plt.show()
    plt.figure(figsize=(6,4))
    sns.scatterplot(x='HorasEstudio', y='Calificacion', data=df)
    plt.title(f'Dispersin: Horas de Estudio vs Calificacin - {nombre_grupo}')
    plt.xlabel('Horas de Estudio')
    plt.ylabel('Calificacin Final')
    X = df[['HorasEstudio']]
    y = df['Calificacion']
    modelo = LinearRegression()
    modelo.fit(X, y)
    y_pred = modelo.predict(X)
    plt.plot(X, y_pred, color='red', label='Lnea de Regresin')
    plt.legend()
    plt.show()
    pendiente = modelo.coef_[0]
    intercepto = modelo.intercept_
    correlacion, p_valor = pearsonr(df['HorasEstudio'], df['Calificacion'])
    print(f"\nEcuacin de la regresin: y = {pendiente:.2f}x + {intercepto:.2f}")
    print(f"Coeficiente de correlacin de Pearson: {correlacion:.3f}")
    print(f"Valor-p: {p_valor:.4f}")
    fuerza = "fuerte" if abs(correlacion) > 0.7 else "moderada" if abs(correlacion) > 0.4 else "dbil"
    direccion = "positiva" if correlacion > 0 else "negativa"
    print(f"Interpretacin: Relacin {fuerza} y {direccion} entre las variables.")
    print("Modelo estadsticamente significativo " if p_valor < 0.05 else " Modelo NO significativo")
```

```python
# Ejecutar anlisis con nombres descriptivos
analizar_grupo("grupo1.csv", "Grupo1")
analizar_grupo("grupo2.csv", "Grupo2")
analizar_grupo("grupo3.csv", "Grupo3")
```
