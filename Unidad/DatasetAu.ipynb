```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
#Cargar el archivo con una variable llamada autos
autos= pd.read_csv("autos.csv", encoding= "ISO-8859-1")
```

```python
#Mostrar los datos estadisticos descriptivos de precios
print("Estadísticas descriptivas:")
autos.price.describe()
```

```python
autos.head()
```

```python
autos.info()
```

```python
#Mostrar el valor minimo de la base de datos de precios
autos.price.min()
```

```python
#Mostrar el valor maximo de la base de datos de precios
autos.price.max()
```

```python
#Mostrar la moda de la base de datos de precios
autos.price.mode()
```

```python
#Mostrar la media de la base de datos de precios
autos.price.mean()
```

```python
#Mostrar la mediana de la base de datos de precios
autos.price.median()
```

```python
#Mostrar la desviacion estandar de la base de datos de precios
autos.price.std()
```

```python
#Histograma
media=autos.price.mean()

plt.pyplot.hist(autos.price, bins=11)
plt.pyplot.title("Histograma Precios de Autos (price)")
plt.pyplot.axvline(media, color='y',linestyle='dashed', linewidth=2, label=f'Media={media:.2f}')
plt.pyplot.xlabel("Precio")
plt.pyplot.ylabel("Frecuencias")
```

```python
# Contar frecuencia de cada categoría
drive_counts = autos['body-style'].value_counts()
# Crear gráfico de barras
plt.pyplot.figure(figsize=(10,5))
plt.pyplot.bar(drive_counts.index, drive_counts.values, color='y',linestyle='dashed')

# Etiquetas y título
plt.pyplot.title("Distribución de Tipos de Vehiculos (body style)")
plt.pyplot.xlabel("Tipo de Vehiculos")
plt.pyplot.ylabel("Frecuencia")
```

```python
# Contar frecuencia de cada categoría
drive_counts = autos['drive-wheels'].value_counts()
# Crear gráfico de barras
plt.pyplot.figure(figsize=(10,10))
plt.pyplot.bar(drive_counts.index, drive_counts.values, color='y',linestyle='dashed')

# Etiquetas y título
plt.pyplot.title("Distribución de Tipos de Tracción (drive wheels)")
plt.pyplot.xlabel("Tipo de Tracción")
plt.pyplot.ylabel("Frecuencia")
```

```python
# Contar frecuencia de cada categoría
drive_counts = autos['engine-location'].value_counts()
# Crear gráfico de barras
plt.pyplot.figure(figsize=(7,7))
plt.pyplot.bar(drive_counts.index, drive_counts.values, color='y',linestyle='dashed')

# Etiquetas y título
plt.pyplot.title("Distribución de Ubicacion del Motor (engine location)")
plt.pyplot.xlabel("Ubicacion del Motor")
plt.pyplot.ylabel("Frecuencia")
```

```python
#Diagrama Caja y Bigotes entre Precio y Tipo de Vehiculo
sns.boxplot(x="price", y="body-style", data=autos)

plt.pyplot.title("Diagrama de Caja: Precio vs Tipo de Vehículo")
plt.pyplot.xlabel("Precio")
plt.pyplot.ylabel("Tipo de Vehículo")
```

```python
#Diagrama Caja y Bigotes entre Precio y Tipo de Traccion
sns.boxplot(x="price", y="drive-wheels", data=autos)

plt.pyplot.title("Diagrama de Caja: Precio vs Tipo de Traccion")
plt.pyplot.xlabel("Precio")
plt.pyplot.ylabel("Tipo de Traccion")
```

```python
#Diagrama Caja y Bigotes entre Precio y Ubicacion del Motor
sns.boxplot(x="price", y="engine-location", data=autos)

plt.pyplot.title("Diagrama de Caja: Precio vs Ubicacion del Motor")
plt.pyplot.xlabel("Precio")
plt.pyplot.ylabel("Ubicacion del Motor")
```
